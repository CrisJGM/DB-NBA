```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('NBA_shoots.csv', sep=',', comment='#')
```

```python
df.columns = df.columns.str.lower()
```

```python
df.columns
```

```python
df.info()
```

```python
# Agrupar por jugador y sumar los valores de 'shot_made' (1 = acierto)
aciertos_por_jugador = (
    df.groupby('player_name')['shot_made']
    .sum()
    .sort_values(ascending=False)
)

# Mostrar el jugador con más y menos aciertos
jugador_top = aciertos_por_jugador.idxmax()
total_aciertos = aciertos_por_jugador.max()

print(f'Jugador con más aciertos: {jugador_top} ({total_aciertos} aciertos)')
print(f'Juagor con menos aciertos: {aciertos_por_jugador.idxmin()} ({aciertos_por_jugador.min()} aciertos)')
```

```python
print(aciertos_por_jugador.head(20))
```

```python
# Agrupar por equipo y sumar los valores de 'shot_made' (1 = acierto)
aciertos_por_team = (
    df.groupby('team_name')['shot_made']
    .sum()
    .sort_values(ascending=False)
)

# Mostrar el equpio con mas y menos aciertos
team_top = aciertos_por_team.idxmax()
total_aciertos = aciertos_por_team.max()

print(f'Equipo con más aciertos: {team_top} ({total_aciertos} aciertos)')
print(f'Equipo con menos aciertos: {aciertos_por_team.idxmin()} ({aciertos_por_team.min()} aciertos)')
```
